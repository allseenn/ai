# Problem

## Two men plays Russian roulette

In revolver with six rounds cylinder there are 2 "bullets" in consecutive chambers ( 2 combat cartridges are in 2 chambers next to each other).

One man **spun the cylinder**, **pulled the trigger** and he is fine, so the **chamber was empty**. 

### What should the second man do: 
**A.** pull the trigger

**B.** spin the cylinder

# Solutions

## Probability

### A. pull the trigger

$$P(A)=\frac{m}{n}$$
- m number of dead events (bullets)
- n number of all possible events

$$P(A)=\frac{2}{5}=0.4$$

### B. spin the cylinder

$$P(B)=\frac{2}{6}\approx 0.33$$


### Answer
Probability of death less in case of **B.** spin the cylinder


## Multitest statistic

### Data

#### cylinder
It's list
[1, 1, 0, 0, 0, 0]
where
- cylinder is list
- 0 empty chamber
- 1 loaded chamber (bullet)
- index[0] assumes chamber with barrel conjanction

#### TESTS
Constant variable. You could set how many test to semulate.

#### SPINS
Constant variable. Maximum number cylinder spins, used for random calculation in function **spin_cylinder()**

#### frags
variable contains all shots during test

#### empties
variable contains all nonshot trigers during test

#### ratio_a
variable contains ratio of **frags** to **empties** and show test statistic of deathes for A. pull the trigger

#### ratio_b
variable contains ratio of **frags** to **empties** and show test statistic of deathes for B. spin the cylinder


### Functions

#### spin_cylinder(lst)

Semulates cylinder spining, randomly shift, not mix bullets in list

#### lucky_pull(lst)

Semulates first man lucky nonshot try, removes any zero from cylinder list

#### pull_trigger(lst)

Checks 0 index of cylinder list, if value is 1 its mean shot, than return 1, overwise return 0



```python
import random

TESTS = 1000000
SPINS = 6

def spin_cylinder(lst):
    shift_amount = random.randint(0, SPINS) 
    shifted_list = lst[shift_amount:] + lst[:shift_amount]
    ##print(shifted_list)
    return shifted_list

def lucky_pull(lst):
    if 0 in lst:
        lst.remove(0)
    return lst

def pull_trigger(lst):
    return lst[0]

# Set of tests for case A. pull the trigger

empties = frags = ratio_a = 0
for _ in range(TESTS):
    cylinder = [1, 1, 0, 0, 0, 0]
    cylinder = spin_cylinder(cylinder)
    cylinder = lucky_pull(cylinder)
    if pull_trigger(cylinder): 
        frags+=1
    else: 
        empties+=1

ratio_a = frags / (empties + frags)

# Set of tests for case B. spin the cylinder
empties = frags = ratio_b = 0
for _ in range(TESTS):
    cylinder = [1, 1, 0, 0, 0, 0]
    cylinder = spin_cylinder(cylinder)
    cylinder = lucky_pull(cylinder)
    cylinder = spin_cylinder(cylinder)
    if pull_trigger(cylinder): frags += 1
    else: empties+=1
        
ratio_b = frags / (empties + frags)

print(f"{ratio_a=} {ratio_b=}")
```

# Conclusions
Statistics from 1 million tests show different results compared to the probability-based solution. 
However, in either case (probability or testing), it is recommended to use the second decision B: spin the cylinder. 
This is because the chances of survival are higher, and the risk of a fatal outcome is lower.
